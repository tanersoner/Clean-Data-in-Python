```python
# Specify indexes for specific rows
indices_to_update = [1, 3] # We want to update lines 2 and 4
# Assign a new categorical value to these rows
df.loc[indices_to_update, 'cleanliness'] = 'Unacceptable'
```

```python
print('Cleanliness: ', df['cleanliness'].unique(), "\n")
```

Let take a look at the output.  Out of the cleanliness columns which one has an inconsistent category. It has an Unacceptable category.

Now I want to drop inconsistent values. I make categories list with 'Clean' 'Unacceptable' 'Average' 'Somewhat clean' 'Somewhat dirty'
 'Dirty' value. 

```python
data = {'cleanliness': ['Clean', 'Average', 'Somewhat clean', 'Somewhat dirty', 'Dirty'],'safety': ['Neutral', 'Very safe', 'Somewhat safe', 'Very unsafe', 'Somewhat unsafe'],'satisfaction': ['Very satisfied', 'Neutral', 'Somewhat satisfied', 'Somewhat unsatisfied', 'Very unsatisfied']}
# Converting data into a pandas DataFrame
categories = pd.DataFrame(data)
categories.head()
```

Again I drop the inconsistent category in the cleanliness column in the dataframe df.

```python
# Find the cleanliness category in airlines not in categories
cat_clean = set(df['cleanliness']).difference(categories['cleanliness'])

# Find rows with that category
cat_clean_rows = df['cleanliness'].isin(cat_clean)

# Print rows with inconsistent category
print(df[cat_clean_rows])
```

```python
# Print rows with consistent categories only
print(df[~cat_clean_rows])
```

Make new dataframe with consistent categories only 

```python
df_cc = df[~cat_clean_rows]
print(df_cc.head())
```

Make a control unique values in cleanliness column

```python
print('Cleanliness: ', df_cc['cleanliness'].unique(), "\n")
```
