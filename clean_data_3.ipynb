```python
SELECT acct_amount, CAST(acct_amount AS INT) AS Account_Amount
FROM 'banking_dirty.csv';
```

```python
SELECT CONCAT(acct_amount, ' Dollars') AS Amount_of_Account
FROM 'banking_dirty.csv';
```

```python
SELECT EXTRACT (MONTH FROM account_opened) AS account_month,
COUNT(*) AS number_orders
FROM 'banking_dirty.csv'
GROUP BY account_month
```

```python
SELECT EXTRACT (MONTH FROM account_opened) AS account_month,
SUM(acct_amount) AS sum_account_amount
FROM 'banking_dirty.csv'
GROUP BY account_month
```
